# 3. Importing misspellings.txt

```python
# # Reading the misspellings file
with open('/content/gdrive/My Drive/nlp/A3/misspellings.txt', 'r',  encoding='utf-8') as f:
    misspellings = f.read()         

# Removing '\n' from the misspellings.txt file
misspellings = [line.strip() for line in misspellings.split('\n')]
  
```

```python
misspellings
```

```python
misspelling = [] # List will contain tuple of (correct,incorrect)

for s in misspellings:
    split_s = s.split('\t')
    prefix = split_s[0].split(',')[0]
    for i in range(1, len(split_s)):
        misspelling.append((prefix, split_s[i]))
```

```python
misspelling
```

```python
# Compute the probabilities of each type of error
insert_prob = defaultdict(int)
delete_prob = defaultdict(int)
transpose_prob = defaultdict(int)
substitute_prob = defaultdict(int)
```

```python
for correct, incorrect in misspelling:
    if len(correct) == len(incorrect):
        # substitute error
        for c, i in zip(correct, incorrect):
            if c != i:
                substitute_prob[(c, i)] += 1

    elif len(correct) == len(incorrect) + 1:
        # delete error
        for i in range(len(incorrect)):
            if correct[i] != incorrect[i]:
                delete_prob[correct[i]] += 1
        delete_prob[correct[-1]] += 1

    elif len(correct) == len(incorrect) - 1:
        # insert error
        for i in range(len(correct)):
            if correct[i] != incorrect[i]:
                insert_prob[(correct[i], incorrect[i])] += 1

    elif len(correct) == len(incorrect) and set(correct) == set(incorrect):
        # transpose error
        for i in range(len(incorrect) - 1):
            if incorrect[i:i+2][::-1] == correct[i:i+2]:
                pair = (incorrect[i], incorrect[i+1])
                transpose_prob[pair] = transpose_prob.get(pair, 0) + 1
    
```

```python
# Normalize the probabilities
insert_prob = {k: v / sum(insert_prob.values()) for k, v in insert_prob.items()}
delete_prob = {k: v / sum(delete_prob.values()) for k, v in delete_prob.items()}
transpose_prob = {k: v / sum(transpose_prob.values()) for k, v in transpose_prob.items()}
substitute_prob = {k: v / sum(substitute_prob.values()) for k, v in substitute_prob.items()}
```
