# <h3>FAIZ</h3>

```python
nlp = spacy.blank('ur')

import csv

with open('faiz.txt', 'rt', encoding= 'utf-8') as f:
  reader = csv.reader(f)
  data = list(reader)
f.close()
```

```python
data
```

```python
len(data)
```

```python
print('no. of Rows:',len(data))
print('no. of Columns:',len(data[0]))
```

```python
def word_tokenization(data):
  list_of_words = []
  for i in range(0, len(data)):
    
    # if the list(row) is not empty([]) do this
    if data[i]:
      # if the list is not empty([' ']) do this
      if data[i][0] != ' ':
        words = nlp(data[i][0])

        for word in words:
          list_of_words.append(word)
  
  return list_of_words
```

```python
words_list = word_tokenization(data)
words_list
```

```python
def sentence_unigram(data):
  list_of_unigrams = []

  for i in range(len(data)):
    # if the list is not empty([]) do this
    if data[i]:
      # if the list is not empty([' ']) do this
      if data[i][0] != ' ':
        sentence = nlp(data[i][0])
        unigrams = [token.text for token in sentence]

        list_of_unigrams.append(tuple(unigrams))
  
  return list_of_unigrams
```

```python
def sentence_bigram(data):
  list_of_bigrams = []

  for i in range(len(data)):
    # if the list is not empty([]) do this
    if data[i]:
      # if the list is not empty([' ']) do this
      if data[i][0] != ' ':
        sentence = nlp(data[i][0])
        bigrams = [token.text + ' ' + sentence[i+1].text for i, token in enumerate(sentence) if i < len(sentence)-1]

        list_of_bigrams.append(tuple(bigrams))
  
  return list_of_bigrams
```

```python
unigrams = sentence_unigram(data)
unigrams
```

```python
bigrams = sentence_bigram(data)
bigrams
```

```python
bigrams[0]
```

```python
len(bigrams)
```

```python
import nltk
from nltk.probability import ConditionalFreqDist
```

```python
cdist = ConditionalFreqDist()

for w in bigrams[0]:
  condition = len(w)
  cdist[condition][w] += 1 

cdist.conditions()


list_of_cdist = []
list_of_cdist.append(cdist[cdist.conditions()[0]])
list_of_cdist.append(cdist[cdist.conditions()[1]])
list_of_cdist.append(cdist[cdist.conditions()[2]])
```
